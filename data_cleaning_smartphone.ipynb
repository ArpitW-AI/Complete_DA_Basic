```python
def res(x):
    if pd.isnull(x):
        return x
    if x<200:
        return 'LOW'
    elif x<300:
        return 'HD'
    elif x<400:
        return 'FULL_HD'
    elif x<500:
        return '2K'
    elif x<600:
        return '4K'
    else :
        return 'ULTRA_HIGH'
```
